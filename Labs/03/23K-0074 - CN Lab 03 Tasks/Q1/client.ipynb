```python
import socket
import os
```

```python
c = socket.socket()
print("socket created correctly.")
```

```python
c.connect(('localhost', 9999))
print("connected to server correctly.")
```

```python
def receive_line(conn):
    data = b''
    
    while not data.endswith(b'\n'):
        chunk = conn.recv(1)
        if not chunk:
            return None
        data += chunk
    return data.decode().strip()
```

```python
while True:
    msg = input("client: ")
    
    if msg.startswith("/file"):
        filename = msg[6:].strip()
        
        if not os.path.exists(filename):
            print("file not found/doesn't exist.")
            continue
        
        filesize = os.path.getsize(filename)
        header = f"FILE|{filename}|{filesize}\n"
        c.sendall(header.encode())
        
        with open(filename, 'rb') as f:
            while True:
                chunk = f.read(1024)
                if not chunk:
                    break
                c.sendall(chunk)
        
        print("file was sent successfully.")
    
    else:
        message = f"TEXT|{msg}\n"
        c.sendall(message.encode())
        
        if msg.lower() == "exit":
            print("closing connection now.")
            break
        
        response = receive_line(c)
        
        if response:
            parts = response.split('|')
            
            if parts[0] == "TEXT":
                print("server: ", parts[1])

c.close()
```
