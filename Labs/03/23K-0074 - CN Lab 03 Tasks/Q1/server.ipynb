```python
import socket
import os
```

```python
def receive_line(conn):
    data = b''
    
    while not data.endswith(b'\n'):
        chunk = conn.recv(1)
        if not chunk:
            return None
        data += chunk
    return data.decode().strip()
```

```python
s = socket.socket()
print("socket created correctly.")
```

```python
s.bind(('localhost', 9999))
```

```python
s.listen(1)
print("socket is listening and waiting for connection..")
```

```python
c, addr = s.accept()
print("connection made with: ", addr)
```

```python
while True:
    header = receive_line(c)
    if not header:
        print("client has beeen disconnected.")
        break
    
    parts = header.split('|')
    
    if parts[0] == "TEXT":
        message = parts[1]
        print("client sent this message: ", message)
        
        if message.lower() == "exit":
            print("client has disconnected.")
            break
        
        reply = input("server: ")
        response = f"TEXT|{reply}\n"
        c.sendall(response.encode())
        
        if reply.lower() == "exit":
            print("server has closed the connection.")
            break
    
    elif parts[0] == "FILE":
        filename = parts[1]
        filesize = int(parts[2])
        
        print(f"server is receiving file: {filename} that has size: {filesize} bytes.")
        
        with open("received_" + filename, "wb") as f:
            rem = filesize
            
            while rem > 0:
                chunk_size = c.recv(min(1024, rem))
                
                if not chunk_size:
                    break
                
                f.write(chunk_size)
                
                rem -= len(chunk_size)
                
        print(f"file saved as : recieved_{filename}")
        
c.close()
s.close()
```
