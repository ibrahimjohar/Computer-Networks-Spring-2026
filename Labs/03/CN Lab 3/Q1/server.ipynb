```python
import socket 
```

```python
s = socket.socket()
print("socket created")
```

```python
s.bind(('localhost', 9999))
```

```python
s.listen(1)
print('waiting for connection...')
```

```python
c, addr = s.accept()
print(f'connected with {addr}') 
```

```python
while True:
    msg_type = c.recv(1024).decode()
    if msg_type == 'TEXT':
        msg = c.recv(1024).decode()
        print('client: ', msg)

        if msg.lower() == 'exit':
            print('client disconnected')
            break

        reply = input('server: ')
        c.send("TEXT".encode())
        c.send(reply.encode())

        if reply.lower() == 'exit':
            print('connection closed by server')
            break

    if msg_type == 'FILE':
        filename = c.recv(1024).decode()
        print(f'receiving file {filename}')
        fileSize = int(c.recv(1024).decode())

        #we need to keep receiving until we have all the bytes so we start with an empty byte obj in python
        file_data = b"" 

        while len(file_data) < fileSize:
            chunk = c.recv(1024)
            file_data += chunk

        with open("received_"+filename, 'wb') as f:
            f.write(file_data)

            print(f'file saved as: recieved_{filename}')

c.close()
s.close()
```
