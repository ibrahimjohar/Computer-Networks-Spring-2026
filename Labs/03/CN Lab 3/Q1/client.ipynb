### Client

```python
import socket 
import os
```

```python
s = socket.socket()
print('socket created')
```

```python
s.connect(('localhost', 9999))
print('connected to server')
```

```python
while True:
    msg = input('client: ')
    if msg.startswith('/file '): 
        filename = msg[6:].strip()

        if not os.path.exists(filename):
            print('file not found')
            continue 

        #sending message type, name and size of file to server
        s.send("FILE".encode())  
        s.send(filename.encode())
        fileSize = os.path.getsize(filename)
        s.send(str(fileSize).encode())
    
        with open(filename, 'rb') as f: #sending file data (opened file in read binary mode)
            s.send(f.read())
    
        print(f'file {filename} sent to server')

    else:
        s.send('TEXT'.encode())
        s.send(msg.encode())

        if msg.lower() == 'exit':
            print('disconnecting...')
            break

        reply_type = s.recv(1024).decode() #recieve reply type

        if reply_type == "TEXT":
            reply = s.recv(1024).decode()
            print("server: ", reply)

            if reply.lower() == 'exit':
                print('server closed connection')
                break

s.close()
        
```
